```python
import matplotlib.pyplot as plt

import pandas as pd
df=pd.read_csv("student-por.csv")
df
```

```python
#continuous data

plt.figure(figsize=(8, 6))
plt.hist(df['age'], bins=17, color='skyblue', edgecolor='black')


plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age Distribution')

plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['sex'], bins=10, color='red', edgecolor='black')


plt.xlabel('Sex')
plt.ylabel('Frequency')
plt.title('Histogram of sex Distribution')

plt.show()
```
